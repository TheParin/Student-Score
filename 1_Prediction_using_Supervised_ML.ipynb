# Prediction using Supervised ML
Predict the percentage of a student based on the no. of study hours. What will be predicted score if a student studies for 9.25 hrs/ day?


![](https://images.pexels.com/photos/4144101/pexels-photo-4144101.jpeg)

The first step is to import the libraries that are required for the implementation of the code operation.  Here we import the Pandas to import and analyze data, NumPy to perform the multi-dimensional operation, and matplotlib to perform graphical plot into the context.

```python
import pandas as pd  
import numpy as np    
import matplotlib.pyplot as plt
import csv
import seaborn as snb
```

The next phase is to load the data into the program to perform the desired operation. 


```python
df = pd.read_csv('student_scores - student_scores.csv')
df
```

Checking for any null values 

```python
df.isnull().sum()
```

Checking for any missing values

```python
df.isna().sum()
```

The next phase is to enter distribution scores and plot them according to the requirement


```python
df.plot(x='Hours', y='Scores', style='o')    
plt.title('Hours vs Score')    
plt.xlabel('Hours')    
plt.ylabel('Score')    
plt.show()  
```

The process of dividing the data into attributes and labels

```python
X = df.iloc[:, :-1].values    
y = df.iloc[:, 1].values  
```

we will be using Scikit Learn's builtin method of train_test_split(), as below

```python
from sklearn.model_selection import train_test_split    
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

Train the algorithm

```python
from sklearn.linear_model import LinearRegression    
regressor = LinearRegression()    
regressor.fit(X_train, y_train)
```

Plotting test data

```python
line = regressor.coef_*X+regressor.intercept_  
plt.scatter(X, y)  
plt.plot(X, line);  
plt.show()  
```

It is visibly clear that the Linear correlation is present
Now, Predicting the scores for the model

```python
y_pred = regressor.predict(X_test)  
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})    
df
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
r2_score = regressor.score(X_test,y_test)
print(r2_score*100,'%')
```

# **What will be predicted score if a student studies for 9.25 hrs/ day?**

```python
hours = [[9.25]]  
Prediction = regressor.predict(hours)  
print("Hours = {}".format(hours))  
print("Prediction Score = {}".format(Prediction[0])) 
```
